```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset CSV
df = pd.read_csv("../data/PROCESSED/train.csv")

# Assuming last 14 columns are disease labels
class_names = df.columns[-14:]
pos_counts = df[class_names].sum().values

# Plot bar chart
plt.figure(figsize=(12,6))
plt.bar(class_names, pos_counts, color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.ylabel("Positive Sample Count")
plt.title("Class Imbalance in ChestX-ray14 Training Set")
plt.tight_layout()
plt.show()
```
